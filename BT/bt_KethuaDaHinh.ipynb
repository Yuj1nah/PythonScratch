```python
#Bài 1. Kế thừa lớp Person
class Person:
    def __init__(self,code,name,birth):
        self.code=format(code,'04d')
        self.name=str(name)
        self.name=self.name.title()
        self.birth=birth
    def chuanhoa(self):
        if self.birth[1]=='/':
            self.birth='0'+self.birth
        if self.birth[4]=='/':
            self.birth=self.birth[0:3]+'0'+self.birth[3:]
        self.name=self.name.title()
class SV(Person):
    def __init__(self,code,name,birth,lop,gpa):
        super().__init__(code,name,birth)
        self.lop=lop
        self.gpa=gpa
    def infor(self):
        print(self.code+' '+self.name +' '+self.birth[0:10]+' '+self.birth[10:]+' '+self.lop+' '+self.gpa)
if __name__=='__main__':
    n=int(input())
    for i in range(n):
        s=SV(i+1,input(),input(),input(),input())
        s.chuanhoa()
        s.infor()
```

```python
#Bài 2. Kế thừa lớp Person và sắp xếp theo tên
class Person:
    def __init__(self,name,birth,locate):
        self.name=str(name).title()
        self.birth=birth
        self.locate=locate
class SV(Person):
    def __init__(self,code,name,birth,locate,lop,gpa):
        self.code=format(code,'04d')
        super().__init__(name,birth,locate)
        self.lop=lop
        self.gpa=gpa
    def chuanhoa(self):
        if self.birth[1]=='/':
            self.birth='0'+self.birth
        if self.birth[4]=='/':
            self.birth=self.birth[0:3]+'0'+self.birth[3:]
    def get_name(self):
        ho,dem,ten=self.name.split()
        return ten
    def get_demho(self):
        ho,dem,ten=self.name.split()
        return ho,dem
    def infor(self):
        print(self.code+' '+self.name+' '+self.birth+' '+self.locate+' '+self.lop+' '+self.gpa)
if __name__=='__main__':
    n=int(input())
    a=[]
    for i in range(n):
        s=SV(i+1,input(),input(),input(),input(),input())
        s.chuanhoa()
        a.append(s)
    a.sort(key=lambda x: (x.get_name(),x.get_demho()))
    for i in a:
        i.infor()
```

```python
# Bài 3. Sinh Viên và Giáo Viên
class Person:
    def __init__(self,name,birth,hometown):
        self.name=str(name).title()
        self.birth=birth
        self.hometown=hometown
    def chuanhoa(self):
        if self.birth[1]=='/':
            self.birth='0'+self.birth
        if self.birth[4]=='/':
            self.birth=self.birth[0:3]+'0'+self.birth[3:]
class CV(Person):
    def __init__(self,code,name,birth,hometown,falcuty,gpa):
        self.code=code
        super().__init__(name,birth,hometown)
        self.falcuty=falcuty
        self.gpa=gpa
    def infor(self):
        print(self.code+' '+self.name+' '+self.birth+' '+self.hometown+' '+self.falcuty+' '+self.gpa)
if __name__=='__main__':
    n=int(input())
    GV=[]
    SV=[]
    for i in range(n):
        s=CV(input(),input(),input(),input(),input(),input())
        s.chuanhoa()
        if s.code[0:2]=='GV':
            GV.append(s)
        else:
            SV.append(s)
    print('DANH SACH GIAO VIEN :')
    for i in GV:
        i.infor()
    print('DANH SACH SINH VIEN :')
    for i in SV:
        i.infor()
```

```python
#Bài 4. Truy vấn sinh viên, giáo viên
class Person:
    def __init__(self,name,birth,country):
        self.name=str(name).title()
        self.birth=birth
        self.country=country
    def chuanhoa(self):
        if self.birth[1]=='/':
            self.birth='0'+self.birth
        if self.birth[4]=='/':
            self.birth=self.birth[0:3]+'0'+self.birth[3:]
    def get_country(self):
        return self.country
    def __str__(self):
        return self.name+' '+self.birth+' '+self.country
class SinhVien(Person):
    def __init__(self,ma,name,birth,country,lop,gpa):
        self.ma=ma
        super().__init__(name,birth,country)
        self.lop=lop
        self.gpa=gpa
    def __str__(self):
        return self.ma+' '+Person.__str__(self)+' '+self.lop+' '+str(self.gpa)
class GiaoVien(Person):
    def __init__(self,ma,name,birth,country,falculty,luong,lop):
        self.ma=ma
        super().__init__(name,birth,country)
        self.falculty=falculty
        self.luong=luong
        self.lop=lop
    def __str__(self):
        return self.ma+' '+Person.__str__(self)+' '+self.falculty+' '+str(self.luong)
if __name__=='__main__':
    n=int(input())
    SV=[]
    GV=[]
    for i in range(n):
        ma=input()
        if ma[0:2]=='GV':
            t=GiaoVien(ma,input(),input(),input(),input(),input(),input())
            t.chuanhoa()
            GV.append(t)
        else:
            s=SinhVien(ma,input(),input(),input(),input(),input())
            s.chuanhoa()
            SV.append(s)
    a=input()
    print('DANH SACH GIAO VIEN CO DIA CHI TAI {} :'.format(a))
    for i in GV:
        if i.get_country()==a:
            print(i)
    print('DANH SACH SINH VIEN CO DIA CHI TAI {} :'.format(a))
    for i in SV:
        if i.get_country()==a:
            print(i)
```

```python
#Bài 5. Sắp xếp theo lương, Gpa
class Person:
    def __init__(self,ma,name,birth,address):
        self.ma=ma
        self.name=str(name).title()
        self.birth=birth
        self.address=address
    def chuanhoa(self):
        if self.birth[1]=='/':
            self.birth='0'+self.birth
        if self.birth[4]=='/':
            self.birth=self.birth[0:3]+'0'+self.birth[3:]
    def get_ma(self):
        return self.ma
    def __str__(self):
        return self.ma+' '+self.name+' '+self.birth+' '+self.address
class SinhVien(Person):
    def __init__(self,ma,name,birth,address,lop,GPA):
        super().__init__(ma,name,birth,address)
        self.lop=lop
        self.GPA=GPA
    def get_GPA(self):
        return self.GPA
    def __str__(self):
        return Person.__str__(self)+' '+self.lop+' '+'{:.2f}'.format(self.GPA)
class GiaoVien(Person):
    def __init__(self,ma,name,birth,address,falculty,luong):
        super().__init__(ma,name,birth,address)
        self.falculty=falculty
        self.luong=luong
    def get_luong(self):
        return self.luong
    def __str__(self):
        return Person.__str__(self)+' '+self.falculty+' '+str(self.luong)
if __name__=='__main__':
    n=int(input())
    SV=[]
    GV=[]
    for i in range(n):
        ma=input()
        if ma[0:2]=='GV':
            s=GiaoVien(ma,input(),input(),input(),input(),int(input()))
            s.chuanhoa()
            GV.append(s)
        else:
            t=SinhVien(ma,input(),input(),input(),input(),float(input()))
            t.chuanhoa()
            SV.append(t)
    GV.sort(key=lambda x: (-x.get_luong(),x.get_ma()))
    SV.sort(key=lambda x: (-x.get_GPA(),x.get_ma()))
    print('DANH SACH GIAO VIEN :')
    for i in GV:
        print(i)
    print('DANH SACH SINH VIEN :')
    for i in SV:
        print(i)
```

```python
# Bài 7. Vehicle
class Vehicle:
    def __init__(self,ma,ten,brand,color):
        self.ma=ma
        self.ten=ten
        self.brand=brand
        self.color=color
    def get_hang(self):
        return self.brand
    def __str__(self):
        return self.ma+' '+self.ten+' '+self.brand+' '+self.color
class XeMay(Vehicle):
    def __init__(self,ma,ten,brand,color,maxspeed,gia):
        super().__init__(ma,ten,brand,color)
        self.maxspeed=maxspeed
        self.gia=gia
    def __str__(self):
        return Vehicle.__str__(self)+' '+self.maxspeed+' '+self.gia
class XeOto(Vehicle):
    def __init__(self,ma,ten,brand,color,maluc,gia):
        super().__init__(ma,ten,brand,color)
        self.maluc=maluc
        self.gia=gia
    def __str__(self):
        return Vehicle.__str__(self)+' '+self.maluc+' '+self.gia
if __name__=='__main__':
    n=int(input())
    OTO=[]
    XM=[]
    for i in range(n):
        ma=input()
        if ma[0:2]=='XM':
            s=XeMay(ma,input(),input(),input(),input(),input())
            XM.append(s)
        else:
            t=XeOto(ma,input(),input(),input(),input(),input())
            OTO.append(t)
    ten=input()
    print('DANH SACH XE HANG '+ten+' :')
    for i in OTO:
        if i.get_hang()==ten:
            print(i)
    for i in XM:
        if i.get_hang()==ten:
            print(i)
```
