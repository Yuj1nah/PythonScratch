```python
#Bài 10. Cân bằng nguyên tố
from math import *
def prime(a):
    if a<2:
        return False
    for i in range(2,isqrt(a)+1):
        if a%i==0:
            return False
    return True
if __name__=='__main__':
    n=int(input())
    a=list(map(int,input().split()))
    for i in range(n):
        left=0
        for j in range(0,i):
            left+=a[j]
        right=0
        for j in range(i+1,n):
            right+=a[j]
        if prime(left) and prime(right):
            print(i,end=' ')
```

```python
#Bài 11. Liệt kê và đếm số fibonacci.
from math import *
def fibo(n):
    if n==0 or n==1:
        return True
    fn2,fn1=0,1
    for i in range(2,100):
        fn=fn2+fn1
        if fn==n:
            return True
        fn2,fn1=fn1,fn
    return False
if __name__=='__main__':
    n=int(input())
    a=list(map(int,input().split()))
    check=0
    for i in a:
        if fibo(i):
            print(i,end=' ')
            check=1
    if check==0:
        print('NONE')
```

```python
#Bài 16. Liệt kê
from math import *
def prime(n):
    if n<2:
        return False
    for i in range(2,isqrt(n)+1):
        if n%i==0:
            return False
    return True
def palin(n):
    tmp=n
    tong=0
    if n<10:
        return True
    while(n!=0):
        tong=tong*10+n%10
        n//=10
    return tong==tmp
def scp(n):
    a=isqrt(n)
    return a*a==n
def sum_digitprime(n):
    tong=0
    while n!=0:
        tong+=n%10
        n//=10
    if prime(tong):
        return True
    return False
if __name__=='__main__':
    n=int(input())
    a=list(map(int,input().split()))
    snt=[x for x in a if prime(x)]
    stn=[x for x in a if palin(x)]
    scph=[x for x in a if scp(x)]
    sumdigitprime=[x for x in a if sum_digitprime(x)]
    print(len(snt),len(stn),len(scph),len(sumdigitprime),sep='\n')
```
